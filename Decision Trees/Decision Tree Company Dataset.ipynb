```python
from sklearn.preprocessing import LabelEncoder
encoder1=LabelEncoder()
data['ShelveLoc'] =encoder1.fit_transform(data['ShelveLoc'])
data['Urban'] =encoder1.fit_transform(data['Urban'])
data['US'] =encoder1.fit_transform(data['US'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.metrics import classification_report
```

```python
x=data.iloc[:,1:]
y=data['Sales']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size =0.33,random_state=10)
```

Criterion according to "Entropy"

```python
model =DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
y_pred =model.predict(x_test)
```

```python
from sklearn import tree
tree.plot_tree(model)
```

```python
fn =['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['Medium', 'High', 'Low']
fig,axes =plt.subplots(nrows=1,ncols=1,figsize =(4,4),dpi =300)
tree.plot_tree(model,feature_names=fn,
               class_names=cn,
               filled= True)
```

```python
y_pred
```

```python
accuracy =accuracy_score(y_test,y_pred)
accuracy
```

```python
count_misclassified =(y_test != y_pred).sum()
count_misclassified
```

```python
print(classification_report(y_test,y_pred))
```
